## **Python while loop**

what is while loop?
### A while loop is a control sturture in Python(and many other proramming languages)that allows you to execute a block of code repeatdly as long as a specific condition is True. Its like giving your programe a task to do repeadly until a certain condition is met.

```python
count=1
while count<=5:
  print(count)
  count +=1
```

```python
count=20
while count>10:
  print(count)
  count -=10
```

```python
count=2
while count<=10:
  print(count)
  count=count+2
```
